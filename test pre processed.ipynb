```python
import pandas as pd
```

```python
testdata=pd.read_excel("Test_set.xlsx")
print(testdata.head(3))
print(testdata.shape)
print("----------------------------------------------------------------")
print(testdata.info())
print("----------------------------------------------------------------")
print(testdata.describe())
print("----------------------------------------------------------------")
print(testdata.columns)
```

```python
Airline = testdata[["Airline"]] #creating dummies features for airline there are only 12 unique airlines
Airline = pd.get_dummies(Airline, drop_first= True)
Airline.head()
```

```python
Source = testdata[["Source"]] #creating dummies features for source there are only 5 unique source
Source = pd.get_dummies(Source, drop_first= True)
Source.head()
```

```python
Destination = testdata[["Destination"]] #creating dummies features for destination there are only 6 unique destination
Destination = pd.get_dummies(Destination, drop_first = True)
Destination.head()
```

```python
datatest = pd.concat([testdata, Airline, Source, Destination], axis = 1) #concatination all 3 dummies dataframe AIRLINE,SOURCE,DESTINATION
datatest.head(3)
```

```python
datatest.drop(["Airline", "Source", "Destination"], axis = 1, inplace = True) # now droping ariline source destination from data frame after creating dummies
datatest.head(3)
```

```python
datatest.drop(["Route", "Additional_Info"], axis = 1, inplace = True) # no use of i think of route and additional info most of them are blank
datatest.head(3)
```

```python
print(testdata["Total_Stops"].value_counts()) # unique stops further converted into 0,1,2,3,4
datatest.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
datatest.head(3)
```

```python
datatest["Journeyday"] = pd.to_datetime(datatest.Date_of_Journey, format="%d/%m/%Y").dt.day
datatest["Journeymonth"] = pd.to_datetime(datatest["Date_of_Journey"], format = "%d/%m/%Y").dt.month
datatest.head(3)
```

```python
datatest.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
datatest["Dephour"] = pd.to_datetime(datatest["Dep_Time"]).dt.hour
datatest["Depmin"] = pd.to_datetime(datatest["Dep_Time"]).dt.minute
datatest.drop(["Dep_Time"], axis = 1, inplace = True)
datatest.head(3)
```

```python
datatest["Arrivalhour"] = pd.to_datetime(datatest.Arrival_Time).dt.hour
datatest["Arrivalmin"] = pd.to_datetime(datatest.Arrival_Time).dt.minute
datatest.drop(["Arrival_Time"], axis = 1, inplace = True)
datatest.head(3)
```

```python
 duration = list(datatest["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:  
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"  
        else:
            duration[i] = "0h " + duration[i]      

durationhours = []
durationmins = []
for i in range(len(duration)):
    durationhours.append(int(duration[i].split(sep = "h")[0]))    
    durationmins.append(int(duration[i].split(sep = "m")[0].split()[-1])) 
```

```python
datatest["Durationhours"] = durationhours
datatest["Durationmins"] = durationmins
datatest.drop(["Duration"], axis = 1, inplace = True)
datatest.head(3)
```

```python
datatest.to_csv("finadatatest.csv")
```
