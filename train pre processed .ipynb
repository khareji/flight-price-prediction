```python
Airline = traindata[["Airline"]] #creating dummies features for airline there are only 12 unique airlines
Airline = pd.get_dummies(Airline, drop_first= True)
Airline.head()
```

```python
Source = traindata[["Source"]] #creating dummies features for source there are only 5 unique source
Source = pd.get_dummies(Source, drop_first= True)
Source.head()
```

```python
Destination = traindata[["Destination"]] #creating dummies features for destination there are only 6 unique destination
Destination = pd.get_dummies(Destination, drop_first = True)
Destination.head()
```

```python
datatrain = pd.concat([traindata, Airline, Source, Destination], axis = 1) #concatination all 3 dummies dataframe AIRLINE,SOURCE,DESTINATION
datatrain.head(3)
```

```python
datatrain.drop(["Airline", "Source", "Destination"], axis = 1, inplace = True) # now droping ariline source destination from data frame after creating dummies
datatrain.head(3)
```

```python
datatrain.drop(["Route", "Additional_Info"], axis = 1, inplace = True) # no use of i think of route and additional info most of them are blank
datatrain.head(3)
```

```python
print(traindata["Total_Stops"].value_counts()) # unique stops further converted into 0,1,2,3,4
datatrain.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
datatrain.head(3)
```

#    # we can use split method to split the day month year hour minute but in pandas there is an method pd.to_datetime through which we will seprate the day month hour minute and store in indiviually
# it will work only where we have format like 24/03/2019,22:20

```python
datatrain["Journeyday"] = pd.to_datetime(datatrain.Date_of_Journey, format="%d/%m/%Y").dt.day
datatrain["Journeymonth"] = pd.to_datetime(datatrain["Date_of_Journey"], format = "%d/%m/%Y").dt.month
datatrain.head(3)
```

```python
datatrain.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
datatrain["Dephour"] = pd.to_datetime(datatrain["Dep_Time"]).dt.hour
datatrain["Depmin"] = pd.to_datetime(datatrain["Dep_Time"]).dt.minute
datatrain.drop(["Dep_Time"], axis = 1, inplace = True)
datatrain.head(3)
```

```python
datatrain["Arrivalhour"] = pd.to_datetime(datatrain.Arrival_Time).dt.hour
datatrain["Arrivalmin"] = pd.to_datetime(datatrain.Arrival_Time).dt.minute
datatrain.drop(["Arrival_Time"], axis = 1, inplace = True)
datatrain.head(3)
```

# but now we have a duration where this method is not working so we have to split the string 

```python
 duration = list(datatrain["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:  
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"  
        else:
            duration[i] = "0h " + duration[i]      

durationhours = []
durationmins = []
for i in range(len(duration)):
    durationhours.append(int(duration[i].split(sep = "h")[0]))    
    durationmins.append(int(duration[i].split(sep = "m")[0].split()[-1])) 
```

```python
datatrain["Durationhours"] = durationhours
datatrain["Durationmins"] = durationmins
datatrain.drop(["Duration"], axis = 1, inplace = True)
```
